```python
dark = cv2.imread("/Users/wangxiang/dark.jpg")
dark = cv2.cvtColor(dark,cv2.COLOR_BGR2RGB)
gray = cv2.imread("/Users/wangxiang/dark.jpg",0)

GRAY = []
for i in range(gray.shape[0]):
    for j in range(gray.shape[1]):
        GRAY.append(gray[i,j])
_=plt.hist(GRAY,bins=256)
```

```python
dark_=cv2.imread("/Users/wangxiang/dark.jpg")
dark_=cv2.cvtColor(dark_,cv2.COLOR_BGR2RGB)

for i in range(3):
    dark_[:,:,i] = cv2.equalizeHist(dark[:,:,i])
```

```python
THRESHOLD = 75

sky =np.zeros(shape=[dark.shape[0],dark.shape[1],3])

for i in range(dark.shape[0]):
    for j in range(dark.shape[1]):
        if gray[i,j]>THRESHOLD:
            sky[i,j,:] = dark[i,j,:]
            dark_[i,j,:] = [0,0,0]
            
```

```python
plt.figure(figsize=(10,10))

plt.subplot(1,2,1)
plt.imshow(sky/255)

plt.subplot(1,2,2)
plt.imshow(dark_/255)
```

```python
plt.figure(figsize=(10,10))

plt.imshow(sky/255+dark_/255)
```
